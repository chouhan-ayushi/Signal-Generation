```python
import numpy as np
import matplotlib.pyplot as plt
#parameter
fs = 10
t= np.linspace(0,1,fs,endpoint=True)#time vector for 1 sec

# generate an analouge signal
frequency=4 # Frequency of the signal in Hz
amplitude=1 # Amplitude of the signal
analog_signal=amplitude*np.sin(2*np.pi*frequency*t)

#generate a discrete signal
sampling_interval = 0.01  # Sampling interval in seconds
t_discrete = np.arange(0, 1, sampling_interval)  # Discrete time points
discrete_signal = amplitude * np.sin(2 * np.pi * frequency * t_discrete)

# Generate a Digital Signal
threshold = 0.1 # Threshold for converting to binary
digital_signal = np.where(discrete_signal >= threshold, 1, 0)

# Plot the signals
plt.figure(figsize=(12, 8))

# Plot Analog Signal
plt.subplot(3, 1, 1)
plt.plot(t, analog_signal, label="Analog Signal", color='b')
plt.title("Analog Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# Plot Discrete Signal
plt.subplot(3, 1, 2)
plt.stem(t_discrete, discrete_signal, label="Discrete Signal", basefmt=" ", linefmt="g-", markerfmt="go")
plt.title("Discrete Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid()
plt.legend()

# Plot Digital Signal
plt.subplot(3, 1, 3)
plt.step(t_discrete, digital_signal, label="Digital Signal", where="mid", color='r')
plt.title("Digital Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude (Binary)")
plt.grid()
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()


```
